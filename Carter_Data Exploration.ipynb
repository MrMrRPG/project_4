```python
# Find Women Chess Masterusernames
wgm_url = "https://api.chess.com/pub/titled/WGM"
wgm_response = requests.get(wgm_url).json()
wgm_usernames = wgm_response["players"]
wgm_usernames
```

```python
responses = []
for wgm in wgm_usernames:
    try:
        chess_data = requests.get(base_url + wgm).json()
        responses.append(chess_data)
        print(f'Chess user {wgm} is from {chess_data["location"]}')
    except:
        print("next")
```
